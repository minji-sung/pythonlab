```python
dic_age = {
    '승은':['권소현','18~22'],
    '선민':['최진실','21~25'],
    '민지':['이다해','23~27']}

age_list = []
min0 = 0
max0 = 0

for i,v in dic_age.items():
    ageList = dic_age[i][1].split("~")
    intList = list(map(int, ageList))
    print(f'{i}는 {v[0]}이고, 최소나이는 {intList[0]}이고 최대나이는 {intList[1]}입니다!')
    min0 += intList[0]
    max0 += intList[1]

print(f'우리팀의 최저평균연령은 {round(min0/len(dic_age))}, 최고연령은 {round(max0/len(dic_age))}')
print(f'우리팀의 얼굴나이 평균은 {round(((min0+max0)/len(dic_age))/2)} 입니다.')
```
