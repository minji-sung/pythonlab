매개변수와 인수

- 매개변수(parameter)와 인수(arguments)는 혼용해서 사용되는 헷갈리는 용어
- 매개변수는 함수에 입력으로 전달된 값을 받는 변수
- 인수는 함수를 호출할 때 전달하는 입력값을 의미

```
def add(a, b): #a, b는 매개변수
    return a+b

print(add(3,4)) #3, 4는 인수
```

````python
#함수의 형태는 입력값과 결과값의 존재 유무에 따라 4가지 유형으로 나뉩니다.

#일반적인 함수
def circle_area(radius):
    area = 3.14 * radius **2
    return area
circle_area(3)

#입력값이 없는 함수
def say():
    return "Hi"
a = say()
a

#결과값이 없는 함수
def add(a, b):
    print("%d, %d의 합은 %d입니다."% (a, b, a+b))
add(1,2)

#입력값도 결과값도 없는 함수
def say_hello():
    print('안녕')
say_hello()
````

````python
#예제> DT스쿨 최저연령은?
alist = {
    '1팀':[20,21,23,33],
    '2팀':[10,10,20,30,25],
    '3팀':[20,30,25,30],
    '4팀':[20,20,20,20],
    '5팀':[30,30,20,20,10],
    '6팀':[20,20,25,30]}

age_li = []
for i,v in alist.items():
    print(f'{i}의 최저연령은{min(v)} 최고연령은 {max(v)}')
    print(f'{i}의 평균은 {sum(v)/len(v)}')
    age_li.extend(v)
    #print(age_li)
print(f'DT스쿨 최저연령은 {min(age_li)}, 최고연령은 {max(age_li)}, 전체평균은 {round(sum(age_li)/len(age_li))}')

````

````python
#예제> 우리팀 캐릭터 사전 만들기

team_dic = {'파':['파이리',10],'이':['이브이',20],'썬':['썬더볼트',30]}
a = input()
b_list = team_dic[a]
print(f'우리 {a}님을 소개합니다. {a}님은 {b_list[0]}를 좋아해요.')
print(f'정신연령은 {b_list[1]}입니다!')
````

````python
#문자열 나누기

#문자열.split()함수는 문자열을 일정한 규칙으로 잘라 리스트로 만드는 함수

text = '문자열 나누기'
li = text.split(sep=' ')
print(li)

#'구분자'.join(리스트) 함수는 리스트에 있는 요소 하나하나를 합쳐서 하나의 문자열로 바꾸어 반환
text2 = ' '.join(li)
text2 
````
