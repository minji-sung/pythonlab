###문자열 나누기

```python
#문자열.split()함수는 문자열을 일정한 규칙으로 잘라서 리스트로 만들어 주는 함수
text = '문자열 나누기'
li = text.split(sep=' ')
print(li)

#'구분자'.join(리스트) 함수는 리스트에 있는 요소 하나하나를 합쳐서 하나의 문자열로 바꾸어 반환
text2 = ' '.join(li)
print(text2)
```
